```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import xarray as xr
```

```python
def plot_comparison(merged_ds: xr.Dataset, var1, var2, output_file=None):
    """Create comparison plot."""
    fig = go.Figure()
    #increase height and width
    fig.update_layout(
        autosize=False,
        width=1800,
        height=1000,
        margin=dict(
            l=50,
            r=50,
            b=100,
            t=100,
            pad=4
        ),
    )
    #Set title to merged_ds["title"]
    fig.update_layout(title=merged_ds.attrs["title"])
    fig.add_trace(go.Scatter(
        x=merged_ds['datetime'],
        y=merged_ds[var1],
        name="SubOcean Depth"
    ))
    
    fig.add_trace(go.Scatter(
        x=merged_ds["datetime"],
        y=merged_ds[var2],
        name="CTD Depth"
    ))
    
    fig.update_layout(yaxis=dict(autorange="reversed"))
    if output_file:
        fig.write_html(output_file)
    return fig
```

```python
merged_ds = xr.open_dataset(r"../data\subocean\forel\CH4\MERGED_forel_SubOceanExperiment2024-07-03T12-47-08.nc_merged.nc")
```

```python
plot_comparison(merged_ds, "Depth_meter_", "Depth_ctd")
```
